```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer ## Words to Sentences then Sentences to Numeric
from sklearn.metrics.pairwise import linear_kernel
```

```python
ds = pd.read_csv("https://raw.githubusercontent.com/rahcoderepo/NLP/refs/heads/main/Amazon_unique_dummy_products.csv")
ds.head()
```

```python
## the following code will:
## analyzer = 'word' -> analyze every word
## ngram_range  = (1,3) -> consolidate 3 words into one unit
## stop_words = 'english' -> eliminate stop words using the english dictionary
tf = TfidfVectorizer(analyzer='word', ngram_range=(1,3), min_df=1, stop_words='english')
tfidf_matrix = tf.fit_transform(ds['description']) ## Only look at the description column

```

```python
cosine_similarities = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
results = {}

for idx, row in ds.iterrows():
  similar_indices = cosine_similarities[idx].argsort()[:-100:-1]
  similar_items = [(cosine_similarities[idx][i], ds['id'][i]) for i in similar_indices]

  results[row['id']] = similar_items[1:]

print('done!')

```

```python
results
```

```python
def item(id):
  return ds.loc[ds['id'] == id]['description'].tolist()[0].split(' - ')[0]

# Just reads the results out of the dictionary
def recommend(item_id, num):
  print("Recommending " + str(num) + " products similar to " + item(item_id) + "...")
  print("-------------")
  recs = results[item_id][:num]
  for rec in recs:
    print("Recommended: " + item(rec[1]) + " (score:" + str(rec[0]) + ")" )
```

```python
recommend(item_id=1, num=5)
```
